```python
df = pd.read_csv("~/Desktop/plots/Selected.csv").T
df
df.rename(columns=df.iloc[0]).drop(df.index[0])
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df['baySeq'].plot(figsize=(15,10),marker='v')
df['DESeq2'].plot(figsize=(15,10),marker='^')
df['edgeR_glmRT'].plot(figsize=(15,10),marker='<')
df['edgeR_glmQLF'].plot(figsize=(15,10),marker='')
df['edgeR_EXACT'].plot(figsize=(15,10),marker='*')
df['limma_trend_eBayes'].plot(figsize=(15,10),marker='*')
df['limma_trend_treat'].plot(figsize=(15,10),marker='P')
df['limma_voom_eBayes'].plot(figsize=(15,10),marker='x')
df['limma_voom_treat'].plot(figsize=(15,10),marker='D')
df['ABSseq'].plot(figsize=(15,10),marker='d')
df['ALDEx2'].plot(figsize=(15,10),marker='|')
df['EBSeq'].plot(figsize=(15,10),marker='o')
df['sSeq'].plot(figsize=(15,10),marker='>')
df['ABSseq-aFold'].plot(figsize=(15,10),marker='>')
plt.xticks(np.arange(len(df.index)), df.index)
plt.legend()
plt.xlim(0,50)
plt.ylabel("Selected genes")
plt.xlabel("Dataset")
#plt.title("Plot of Selected")
plt.show()
```
